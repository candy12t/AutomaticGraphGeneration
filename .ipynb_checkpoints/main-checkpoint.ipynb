```python
#  ライブラリをimport
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
from matplotlib import rcParams
import os
```

```python
#  フォント、スタイルを日本語に設定
mpl.style.use('default')
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'Noto Sans CJK JP']
```

```python
# csvを読み込み
def get_csv(filename):
    return pd.read_csv(filename, encoding='utf-8')
```

```python
df = get_csv( 'ex3_Kp100_Kd10_O=3.csv')
```

```python
df.head(3)
```

```python
#  各カラムを変数に置く
time = df['#time[sec]']
thr = df[' thr[rad]']
th = df[' th[rad]']
d_th = df[' d_th[rad/sec]']
dd_th = df[' dd_th[rad/sec^2]']
```

```python
#  描画の初期設定
def create_figure():
    # 軸を内側に
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
```

```python
def plot():
    create_figure()
    fig = plt.figure(dpi=300)
    ax = fig.add_subplot(111)
    plt.xticks(np.arange(0, 5.5, step=0.5))
    ax.set_xlim(0, 5)
    ax.set_ylim(0, 2.6)
    plt.grid()
    plt.xlabel('時間 [sec]')
    plt.ylabel('回転角度 [θ]')
    plt.plot(time,th,  label='回転角度')
    plt.plot(time, thr,  linestyle='dashed', label='目標回転角度')
    plt.legend()
    plt.savefig('test.png')

plot()
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_subplot(111)
plt.xticks(np.arange(0, 5.5, step=0.5))
ax.set_xlim(0, 5)
plt.grid()
plt.xlabel('時間 [sec]')
plt.ylabel('角加速度 [rad/sec^2]')
plt.plot(time, d_th)
plt.savefig('課題2/おもり3個.png')
```
